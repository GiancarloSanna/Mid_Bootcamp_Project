# Importing the main dataset

```python
status = pd.read_excel("Raw Data/Legal Status.xlsx", sheet_name = "Tabelle1")
status
```

Convert columns to snakecase

```python
cols = []
for column in status.columns:
    cols.append(column.lower())
status.columns = cols

status = status.rename(columns={'in-person':'in_person'})
```

Thats a quick one, ? means the according type of spors betting is legal, everything else means basically not legal.
We map it with 0 for not legal and 1 for legal.

```python
def legal(x):
    if x == '?':
        return 1
    return 0
```

```python
status['online'] = status['online'].apply(legal)
status['in_person'] = status['in_person'].apply(legal)
```

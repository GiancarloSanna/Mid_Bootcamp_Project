### The structure
We can see, that we have the description of the state and what kind of number is displayed all in one column.

The first three rows are like a header/description, we don't need that.

I come to the conclusion, that cleaning this dataframe would take more time than creating a new one, extracting the information we need.


We basically want to have a dataframe with the following columns:

```python
columns = ['year', 'state', 'population', 'poverty','unemployed']
```

```python
povertydata = []
```

```python
# Beginning in line 4 and last on line 361 every 7th line on the dataframe has the statename, we begin from here
# The whole file is for 2018
year = '2018'

for row in range(4,len(poverty['Unnamed: 0']),7):
    povertydata.append([
        2018,
        poverty.iloc[row,0],
        poverty.iloc[row+2,1],
        poverty.iloc[row+4,1],
        poverty.iloc[row+2,36]
    ])
    
```

```python
# No we have to repeat this process for the years 2019-2021
for year in ['2019','2020','2021']:
    poverty = pd.read_excel("Raw Data/Poverty/Poverty"+year+".xlsx", sheet_name = "Data")
    for row in range(4,len(poverty['Unnamed: 0']),7):
        povertydata.append([
        year,
        poverty.iloc[row,0],
        poverty.iloc[row+2,1],
        poverty.iloc[row+4,1],
        poverty.iloc[row+2,36]
        ])
```

# Data Cleaning

First we transfer our array into a dataframe

```python
poverty_df = pd.DataFrame(povertydata,columns = columns)
poverty_df
```

We only have to remove the commas and convert the numbers to integers.

```python
def num_clean(x):
    return int(x.replace(",",""))
```

```python
for col in ['population','poverty','unemployed']:
    poverty_df[col] = poverty_df[col].apply(num_clean)
```
